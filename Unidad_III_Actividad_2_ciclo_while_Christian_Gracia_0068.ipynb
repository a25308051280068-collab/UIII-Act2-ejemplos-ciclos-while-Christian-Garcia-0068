### Acumulador de puntos

```python
#1. Acumulador de puntos con límite (Aritméticos y Relacionales)
#Este código solicita números al usuario y los suma hasta que se alcanza un tope o el usuario decide detenerse con un valor específico.
total = 0
continuar = True
while total < 50 and continuar:
    puntos = int(input("Ingresa puntos a sumar (0 para salir): "))
    if puntos == 0:
        continuar = False
    else:
        total += puntos
        print(f"Total acumulado: {total}")

print("Fin del juego.")
```

### Validador de rango

```python
#2. Validador de rango con reintentos (Lógicos y Relacionales)
#Este ejemplo obliga al usuario a ingresar un número que esté estrictamente dentro de un rango permitido, utilizando lógica de negación.
numero = -1
while not (10 <= numero <= 20):
    numero = int(input("Ingresa un número entre 10 y 20: "))
    if not (10 <= numero <= 20):
        print("¡Fuera de rango! Intenta de nuevo.")

print(f"Número {numero} aceptado.")
```

### Buscador de multiplos

```python
#3. Buscador de múltiplos específicos (Aritméticos, Lógicos y Relacionales)
#Este programa pide números y solo se detiene cuando el usuario ingresa un número que es múltiplo de 3 y de 5 al mismo tiempo.
entrada = 1
while entrada % 3 != 0 or entrada % 5 != 0:
    entrada = int(input("Ingresa un número (el ciclo para con un múltiplo de 3 y 5): "))
    if entrada % 3 != 0 or entrada % 5 != 0:
        print("Ese no es múltiplo de ambos. Sigue intentando.")

print(f"¡Correcto! {entrada} es múltiplo de 3 y 5.")
```

### Control de acceso

```python
#4. Control de acceso con intentos limitados (Lógicos y Relacionales)
#Este programa simula un sistema de seguridad que permite un máximo de 3 intentos para ingresar una clave correcta, utilizando una combinación de condiciones lógicas

intentos = 3
acceso_concedido = False

while intentos > 0 and not acceso_concedido:
    clave = input("Introduce la clave de acceso: ")
    if clave == "secreto123":
        acceso_concedido = True
        print("Acceso permitido.")
    else:
        intentos -= 1
        print(f"Clave incorrecta. Intentos restantes: {intentos}")

if not acceso_concedido:
    print("Cuenta bloqueada por seguridad.")
```

### simulador de cajero

```python
#5. Simulador de cajero con validación (Aritméticos, Lógicos y Relacionales)
#Este ejemplo permite retirar dinero de un saldo inicial mientras el usuario lo desee y existan fondos suficientes, validando que el monto sea positivo

saldo = 1000
continuar = "s"

while saldo > 0 and continuar == "s":
    retiro = int(input(f"Saldo actual: ${saldo}. ¿Cuánto desea retirar? "))
    if retiro > 0 and retiro <= saldo:
        saldo -= retiro
        print(f"Retiro exitoso. Nuevo saldo: ${saldo}")
    else:
        print("Error: Monto inválido o saldo insuficiente.")

    if saldo > 0:
        continuar = input("¿Desea realizar otra operación? (s/n): ")

print("Gracias por usar el cajero.")
print("programa realizad por Christian Garcia 0068")
```
